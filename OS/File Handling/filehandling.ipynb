    
    r ---> read file
    w ----> write file(overwrite)
    a -----> append (adds in last)

| Mode   | Meaning                         |
| ------ | ------------------------------- |
| `"r"`  | Read file                       |
| `"w"`  | Write file (overwrite old data) |
| `"a"`  | Append data at end              |
| `"x"`  | Create new file                 |

```python
file = open("file1.txt",'r')
```

```python
# file.read()  -----> reads entire file
# Whenever you open a file, Python places an invisible pointer (cursor) at the very beginning (index 0).
#  When you call a read method, that cursor moves forward.
file.read()
# This command reads the entire file from start to finish.
# Because it read everything, the cursor travels all the way to the very end of the file (after the last word of the last line).
```

```python
file.seek(0)                 # Reset cursor back to the very beginning (index 0)
```

```python
first_line = file.readline()                #read only first line
print(first_line)
```

```python
file.readlines()           #read multiple lines of the file
```

```python
file.close()
```

```python
with open('file1.txt') as file:      #with works as a loop
    print(file.read())
```

```python
file = open('file1.txt','w')        # w mode is used for write, it overwrites the data of the file 
file.write("I am from Rajasthan")
file.close()
file = open('file1.txt','r')
file.read()
```

```python
file = open('file2.txt','w')    #if the file is not there already it create that
file.write("I am from Rajasthan")
file.close()
file = open('file1.txt','r')
file.read()
```

```python
file = open("file1.txt",'a')             #it add the text in the end without overwriting previous
file.write("This is file handling")
file.close()
```

```python
open('file1.txt','r').read()
```
