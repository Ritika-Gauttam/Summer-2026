## Exception Handling in Python

used to handle run time error


| Concept                 | Definition                                        | Syntax/Keyword                    | Purpose                             | Example                                |
| ----------------------- | ------------------------------------------------- | --------------------------------- | ----------------------------------- | -------------------------------------- |
| **Exception**           | An error that occurs during program execution     | —                                 | Indicates abnormal program behavior | `10/0` → `ZeroDivisionError`           |
| **try**                 | Block containing code that may cause an exception | `try:`                            | Monitors code for errors            | `try: result = 10/0`                   |
| **except**              | Block that handles an exception                   | `except:`                         | Prevents program from crashing      | `except: print("Error")`               |
| **Specific Exception**  | Handles a particular type of error                | `except ZeroDivisionError:`       | Provides customized error handling  | Handle division by zero separately     |
| **Multiple Exceptions** | Handles more than one exception type              | `except (ValueError, TypeError):` | Reduces code duplication            | Handle invalid input and type errors   |
| **else**                | Executes only if no exception occurs              | `else:`                           | Runs success code                   | Print result after successful division |
| **finally**             | Executes whether an exception occurs or not       | `finally:`                        | Cleanup operations                  | Close files, database connections      |
| **raise**               | Manually generates an exception                   | `raise ExceptionType()`           | Enforces custom error conditions    | `raise ValueError("Invalid Age")`      |


| Situation                           | Executed Blocks            |
| ----------------------------------- | -------------------------- |
| No exception occurs                 | `try → else → finally`     |
| Exception occurs and is handled     | `try → except → finally`   |
| Exception occurs and is not handled | `try → Program Terminates` |


| Exception             | Description                            |

```python
class InvalidPassError(Exception):
    pass

try:
    password = input("Enter password:")
    if(len(password)<8):
        raise InvalidPassError("Password should have atleast 8 characters")
    special_char = ["@","!","$"]
    found_special = False
    for char in password:
        for spec in special_char:
            if char == spec:
                found_special = True
                break             
        if found_special:
            break 
    if not found_special :
        raise InvalidPassError("Password must contain a special character")

    print("Password set successfully")

except InvalidPassError as e:
    print(e)

```
