```python
plt.plot([1,2,3,4])
# Without plt.show(), plt.plot() or anything plt. returns a Python list containing actual line objects,
# not just a memory address.When you see something like [<matplotlib.lines.Line2D at 0x7f8b1c2d3e40>]
#  printed on your screen, that text is just a description of the object.
# matplotlib.lines.Line2D: This is the specific Python class type
# at 0x7f8b1c2d3e40: This is the hexadecimal memory address where that specific line object is 
# currently stored in your computer's RAM.
# plt. returns is a Tool, Not Just an Address
```

```python
x = np.random.normal(170 ,10,250)
x
# This code generates an array of 250 random numbers following a normal (bell-shaped) distribution curve.
# Breakdown of the Code
# np.random.normal: Calls the NumPy normal distribution function.
# 170: Sets the average (mean) value.
# 10: Sets the standard deviation (spread).
# 250: Generates exactly 250 data points.
# x: Stores these numbers inside variable x.
```

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
x = np.array([1,3,5,7])
plt.pie(x)  
# plt.pie() only takes one main data argument,Passing two arrays causes an error
plt.show() 
```

```python
x = np.array([1,3,5,7])
plt.pie(x)
```

```python
x = np.array([1,3,5,7])
label = ['apple','mango','banana','pomogranate']
plt.pie(x,labels=label)
plt.show()
```

```python
x = np.array([1,3,5,7])
label = ['apple','mango','banana','anar']
plt.pie(x,labels=label,explode = [0,0.2,0,0])    #To take out piece use explode
plt.show()
```

```python
x = np.array([1,3,5,7])
label = ['apple','mango','banana','anar']
plt.pie(x,labels=label,explode = [0,0.2,0,0],shadow = True)
plt.show()
```

```python
x = np.array([1,3,5,7])
label = ['apple','mango','banana','anar']
colr = ['blue','pink','green','red']
plt.pie(x,labels=label,explode = [0,0.2,0,0],shadow = True, colors=colr)
plt.show()
```

```python
x = np.array([1,3,5,7])
label = ['apple','mango','banana','anar']
plt.pie(x,labels=label,explode = [0,0.2,0,0],shadow = True,autopct='%1.1f%%')
plt.show()
```

```python
x = np.array([1,3,5,7])
label = ['apple','mango','banana','grapes']
plt.pie(x,labels=label,autopct='%1.1f%%',explode = [0,0.2,0,0],shadow = True)
plt.legend()
plt.title("Fruits")
plt.show()
```
