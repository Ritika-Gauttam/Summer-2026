# Numpy
NumPy is a powerful Python library used for:

1. Fast numerical computations
2. Working with arrays and matrices
3. Mathematical functions
4. Scientific computing and data analysis

**The name stands for Numerical Python.**

| Topic                     | Description                                                  | Example                   |
| ------------------------- | ------------------------------------------------------------ | ------------------------- |
| **NumPy**                 | A Python library used for numerical and scientific computing | `import numpy as np`      |
| **Purpose**               | Used for fast array operations and mathematical calculations | Data analysis, ML         |
| **Main Feature**          | Provides powerful multidimensional arrays                    | `np.array()` --> to create array             |
| **Installation**          | Install using pip                                            | `pip install numpy`       |
| **Import Statement**      | Import NumPy with alias `np`                                 | `import numpy as np`      |
| **Array Creation**        | Create NumPy arrays                                          | `np.array([1,2,3])`       |
| **Zeros Function**        | Creates array filled with 0                                  | `np.zeros((2,2))`         |
| **Ones Function**         | Creates array filled with 1                                  | `np.ones((2,2))`          |
| **Arange Function**       | Creates array in a range                                     | `np.arange(0,10,2)`       |
| **Random Values**         | Generates random numbers                                     | `np.random.rand(3)`       |
| **Shape**                 | Returns dimensions of array                                  | `arr.shape`               |
| **Reshape**               | Changes array dimensions                                     | `arr.reshape(2,3)`        |
| **Indexing**              | Access specific element                                      | `arr[0]`                  |
| **Slicing**               | Access multiple elements                                     | `arr[1:4]`                |
| **Addition**              | Element-wise addition                                        | `a + b`                   |
| **Multiplication**        | Element-wise multiplication                                  | `a * b`                   |
| **Mean**                  | Finds average value                                          | `np.mean(arr)`            |
| **Sum**                   | Finds total sum                                              | `np.sum(arr)`             |
| **Maximum Value**         | Finds largest element                                        | `np.max(arr)`             |
| **Minimum Value**         | Finds smallest element                                       | `np.min(arr)`             |
| **Standard Deviation**    | Measures spread of data                                      | `np.std(arr)`             |
| **2D Array (Matrix)**     | Array with rows and columns                                  | `np.array([[1,2],[3,4]])` |
| **Advantages of NumPy**   | Faster and more memory efficient than Python lists           | Useful in large datasets  |
| **Applications**          | Used in AI, ML, Data Science, Image Processing               | Neural networks           |
| **Libraries Using NumPy** | Base library for many ML/data tools                          | Pandas, TensorFlow        |


| Type of Array              | Description                             | Example                                   |
| -------------------------- | --------------------------------------- | ----------------------------------------- |
| **1D Array**               | Contains only one row of elements       | `np.array([1,2,3,4])`                     |
| **2D Array**               | Contains rows and columns (matrix form) | `np.array([[1,2],[3,4]])`                 |
| **3D Array**               | Collection of 2D arrays                 | `np.array([[[1,2],[3,4]],[[5,6],[7,8]]])` |
| **Multidimensional Array** | Array with more than 3 dimensions       | `np.array([[[[1]]]])`                     |

```python


arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([2, 4, 6, 8])

print("Array 1 :", arr1)
print("Array 2 :", arr2)


# Multiplication
print("\nMultiplication (*)")
print("arr1 * arr2 =", arr1 * arr2)

# Division
print("\n Division (/)")
print("arr1 / arr2 =", arr1 / arr2)

# Addition
print("\n Addition (+)")
print("arr1 + arr2 =", arr1 + arr2)

# Subtraction
print("\n Subtraction (-)")
print("arr1 - arr2 =", arr1 - arr2)

# Positive Sign with Multiplication
print("\nMultiplication with Unary Plus (* +)")
print("arr1 * +arr2 =", arr1 * +arr2)

# Floor Division
print("\n Floor Division (//)")
print("arr1 // +arr2 =", arr1 // +arr2)

print("\nAll operations completed successfully!")
```
