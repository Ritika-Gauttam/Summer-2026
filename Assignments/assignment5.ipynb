### Question 1
Write a program to check whether a number is:
● Positive
● Negative
● Zero

```python
num = int(input("Enter a number to check: "))

if num > 0:
    print(num , "is a positive number")
elif num < 0 :
    print(num , "is a negative number")
else :
    print("Your entered number is zero")
```

### Question 2
Write a program to check whether a number is:
● Even
● Odd

```python
num = int(input("Enter a number to check: "))

if num % 2 == 0 :
    print(num , "is an even number")
else:
    print(num , "is an odd number")
```

### Question 4
Write a program to check the largest number among three numbers
using if-elif-else.

```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))

if num1 >= num2 and num1 >= num3:
        print(num1,"is the largest number")
elif num2 >= num1 and num2 >= num3:
        print(num2 , "is the largest number")
else :
    print(num3 , "is the largest number")
```

Write a program using Nested if statement for ATM withdrawal.
Conditions:
● First check PIN
● Then check balance
● If both are correct → Print "Transaction Successful"
● Otherwise print proper message.

```python
correct_PIN = 1972
entered_PIN = int(input("Enter ATM PIN :"))
balance = 500000

if entered_PIN == correct_PIN :
    withdraw = int(input("Enter Amount to withdraw"))
    if balance >= withdraw :
        print("Transaction Successful")
    else:
        print("Insufficient Balance")

else :
    print("Incorrect PIN")
```
