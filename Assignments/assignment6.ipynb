**Question 1.** Create a class NumberSeries that uses a loop to print numbers from 1 to 100.

```python
class NumberSeries:
    def print_num(self):
        for num in range(1,101):
            print(num)

# We can access class properties by creating an object
o1 = NumberSeries()
o1.print_num()
```

**Question 2.**  Create a class EvenOddChecker that checks and prints whether numbers
from 1 to 50 are even or odd using a loop.

```python
class EvenOddChecker:
    def checker(self):
        for x in range(1,51):
            if x % 2 == 0 :
                print(x , "is even")
            else:
                print(x ,"is odd")
            
obj1 = EvenOddChecker()
obj1.checker()
```

**Question 3.** Create a class MultiplicationTable with a function to print the
multiplication table of a user-given number.

```python
class MultiplicationTable:
    def table(self,num):
        for i in range(1,11):
            print(num , "x" , i ,"=" ,num * i)

n = int(input("Enter a number : "))
obj1 = MultiplicationTable()
obj1.table(n)
```

**Question 4.** Create a class FactorialCalculator with a function to calculate the
factorial of a number using a loop.

```python
class FactorialCalculator:
    def factorial(self,num):
        fact = 1
        for i in range(1 , num+1):
            fact = fact * i
        print(fact)    

n = int(input("Enter a Number : "))
obj1 = FactorialCalculator()
obj1.factorial(n)
```

**Question 5.** Create a class PrimeNumberChecker with a function to check whether a
number is prime or not.

```python
class PrimeNumberChecker:
    def prime(self,num):
        for i in range(2,num):
            if num % i == 0:
                print(num , "is not a prime number")
                break ; 
        else :
            print(num, "is a Prime Number")

obj1 = PrimeNumberChecker()
obj1.prime(230)
```

**Question 8.** Create a class Employee using a constructor (__init__) to store
employee details and create a function to calculate yearly salary.

```python
class Employee:

    def __init__(self,name, emp_id ,monthly_salary):
        self.name = name
        self.emp_id= emp_id
        self.monthly_salary = monthly_salary

    def yearly_salary(self):
        yearly= self.monthly_salary * 12 
        print("Employee Name:", self.name)
        print("Employee ID:", self.emp_id)
        print("Yearly Salary:",yearly)


e1 = Employee("Ritika",19,1200000)
e1.yearly_salary()
```

**Question 10.** Create three classes Vehicle, Car, and SportsCar using multilevel
inheritance and display information of a sports car.

```python
class Vehicle:
    company = "BMW"

    def vehicle_info(self):
        print("Company:",self.company)

class Car(Vehicle):
    model = "M4"
    
    def car_info(self):
        print("Model:",self.model)

class SportsCar(Car):
    speed = "250 km/h"

    def SportsCar_info(self):
        print("Speed:",self.speed)


obj = SportsCar()
obj.vehicle_info()
obj.car_info()
obj.SportsCar_info()
```

**Question 11.** Create a class BankAccount with functions:
● deposit money
● withdraw money
● check balance

```python
class BankAccount:
    balance = 0
    
    def deposit_money(self,amount):
        if amount > 0:
            self.balance += amount
            print("Rs.", amount , "deposited successfully.")
        else:
            print("Invalid Amount")
        
    def withdraw_money(self,amount):
        if amount > 0 and amount <= self.balance:
            print("Rs." , amount , "withdrawn successfully.")
            self.balance -= amount
        else :
            print("Insufficient Balance")

    def check_balance(self):
        print("Your Current balance is Rs.",self.balance)


Person1 = BankAccount()
Person1.deposit_money(2000)
Person1.withdraw_money(5000)
Person1.check_balance()

```
