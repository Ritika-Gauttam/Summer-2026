**Question 2** Create a class PrimeChecker with a function to check whether a number is
prime or not.

```python
class PrimeChecker:
    def __init__(self,num):
        self.num = num
        self.checker()

    def checker(self):
        if self.num <= 1:
            print("Not Prime Number")

        else :
            for x in range(2,self.num):
                if (self.num%x==0):
                    print(f"{self.num} is not a Prime Number")
                    break
            else:
                print(f"{self.num} is a Prime Number")


obj1 = PrimeChecker(22)
obj2 = PrimeChecker(2)
```

**Question 5** Create a class Calculator with functions:
● add()
● subtract()
● multiply()
● divide()

```python
class Calculator:
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2
        self.operation()

    def operation(self):
        option = input("Enter operation you want to perform (+ , - , * , /):")
        if option == "+":
            self.add()
        elif option == "-":
            self.subtract()
        elif option == "*":
            self.multiply()
        elif option == "/":
            self.divide()
        else :
            print("Invalid Operation")
        
    def add(self):
        print(self.num1 + self.num2)

    def subtract(self):
        print(self.num1 - self.num2)

    def multiply(self):
        print(self.num1 * self.num2)

    def divide(self):
        if self.num2 == 0:
            print("Can't divide by zero")
        else :
            print(self.num1 / self.num2)


num1 = int(input("Enter first Number:"))
num2 = int(input("Enter second Number:"))
calc = Calculator(num1,num2)
    
```


**Question 6** Create a parent class Person and child class Employee using single
inheritance. Display employee details.


```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
class Employee(Person):
    def __init__(self,name,age,emp_id):
        super().__init__(name,age)
        self.emp_id = emp_id
    
    def display(self):
        print("Employee Name :",self.name)
        print("Employee age :",self.age)
        print("Employee ID :",self.emp_id)

emp1 = Employee("Ritika",19,"INFY1001",)
emp1.display()
```

**Question 7** Create a class BankAccount with functions:
● deposit()
● withdraw()
● check_balance()

```python
class BankAccount:

    def __init__(self,balance):
        self.balance = balance

    def deposit(self,amount):
        if(amount > 0):
            self.balance += amount
            print("Rs.", amount,"desposited successfully")
        else:
            print("Please deposit a valid amount")

    def withdraw(self,amount):
        if (self.balance >= amount):
            self.balance -= amount
            print("Rs.",amount, "withdrawn successfully")
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print("Your Current Balance is Rs.",self.balance)

acc1 = BankAccount(7000)
acc1.deposit(2000)
acc1.withdraw(3000)
acc1.check_balance()
```
