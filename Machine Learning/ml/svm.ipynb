### SVM ----> Support vector machine

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC , SVR

1. SVM Overview

| Feature        | Description                                  |
| -------------- | -------------------------------------------- |
| Full Form      | Support Vector Machine                       |
| Type           | Supervised Learning                          |
| Tasks          | Classification & Regression                  |
| Main Goal      | Find optimal hyperplane with maximum margin  |
| Key Idea       | Maximum margin = Better generalization       |
| Works Well For | Small-Medium datasets, High-dimensional data |


2. Important Terminologies

| Term            | Meaning                                      |
| --------------- | -------------------------------------------- |
| Hyperplane      | Decision boundary                            |
| Margin          | Distance between boundary and nearest points |
| Support Vectors | Nearest points to hyperplane                 |
| Kernel          | Converts data into higher dimensions         |
| Hard Margin     | No misclassification allowed                 |
| Soft Margin     | Some misclassification allowed               |


3. SVM Working Algorithm

| Step | Description                           |
| ---- | ------------------------------------- |
| 1    | Take training data                    |
| 2    | Find all possible hyperplanes         |
| 3    | Calculate margins                     |
| 4    | Select hyperplane with maximum margin |
| 5    | Identify support vectors              |
| 6    | Classify new data points              |


4. Types of SVM

| Type           | Use Case                |
| -------------- | ----------------------- |
| Linear SVM     | Linearly separable data |
| Non-Linear SVM | Complex/non-linear data |


5. Kernels and When to 

| Kernel     | Use When                                | Speed       |
| ---------- | --------------------------------------- | ----------- |
| Linear     | Text data, high-dimensional data        | Fastest     |
| Polynomial | Curved boundaries, feature interactions | Medium      |
| RBF        | Most non-linear datasets                | Medium      |
| Sigmoid    | Neural-network-like behavior            | Rarely Used |

Most Commonly Used Kernel

    RBF Kernel

## Advantages of SVM and Disadvantages of SVM

| Advantages                                       | Disadvantages                   |
| ------------------------------------------------ | ------------------------------- |
| High accuracy                                    | Slow on large datasets          |
| Works in high-dimensional data                   | Difficult parameter tuning      |
| Good for small datasets                          | Memory intensive                |
| Handles non-linear data using kernels            | Less interpretable              |
| Less overfitting due to maximum-margin principle | Sensitive to noise and outliers |


### SVC vs SVR

| Feature      | SVC                       | SVR                      |
| ------------ | ------------------------- | ------------------------ |
| Full Form    | Support Vector Classifier | Support Vector Regressor |
| Problem Type | Classification            | Regression               |
| Output       | Class Label               | Numerical Value          |
| Example      | Spam Detection            | House Price Prediction   |
| sklearn      | `SVC()`                   | `SVR()`                  |


### Linear Regression vs SVR

| Feature                  | Linear Regression | SVR                 |
| ------------------------ | ----------------- | ------------------- |
| Data Relationship        | Linear            | Linear + Non-linear |
| Speed                    | Fast              | Slow                |
| Complexity               | Simple            | Complex             |
| Interpretability         | Easy              | Difficult           |
| Large Dataset            | Better            | Not ideal           |
| Small Dataset            | Good              | Excellent           |
| Accuracy on Complex Data | Lower             | Higher              |
