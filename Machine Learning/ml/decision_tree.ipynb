### Decision Tree

A Decision Tree is a supervised machine learning algorithm that can be used for both Classification and Regression problems.

### Terminologies
| Term           | Definition                                                      |
| -------------- | --------------------------------------------------------------- |
| Root Node      | The topmost node of the tree where splitting starts.            |
| Splitting      | Process of dividing a node into sub-nodes based on a condition. |
| Branch/Subtree | Section of the tree formed after a split.                       |
| Edge           | Connection between two nodes.                                   |
| Decision Node  | A node that further splits into child nodes.                    |
| Leaf Node      | Terminal node that gives the final prediction/output.           |


### Advantages

| Advantage                               | Description                                          |
| --------------------------------------- | ---------------------------------------------------- |
| Easy to Understand                      | Can be visualized and interpreted easily.            |
| No Feature Scaling Required             | Works directly on raw data.                          |
| Handles Non-linear Data                 | Captures complex relationships.                      |
| Works for Classification and Regression | Can predict categories as well as continuous values. |


### Disadvantages

| Disadvantage                         | Description                                                   |
| ------------------------------------ | ------------------------------------------------------------- |
| Overfitting                          | May memorize training data and perform poorly on unseen data. |
| Sensitive to Data Changes            | Small changes in data can create a different tree.            |
| Prone to Errors with Imbalanced Data | Classes with fewer samples may not be predicted well.         |


### CART (Classification and Regression Tree)

| Problem Type   | Output                                                        |
| -------------- | ------------------------------------------------------------- |
| Classification | Predicts class labels (Yes/No, Spam/Not Spam, etc.)           |
| Regression     | Predicts continuous values (Price, Temperature, Salary, etc.) |


CART generally uses the Gini Index for classification and Mean Squared Error (MSE) for regression.

### Information Gain


Information Gain measures how much uncertainty is reduced after splitting a node.

    Information Gain=E(Parent)−Weighted Average of E(Children)

Where:

• E = Entropy

• Parent = Original node

• Children = Nodes formed after splitting

Higher Information Gain ⇒ Better Split


### Entropy

Entropy measures impurity or randomness in the data.

    Entropy=−∑pi (log2(pi))

#### Entropy Range
| Entropy | Meaning                                     |
| ------- | ------------------------------------------- |
| 0       | Pure node (all samples belong to one class) |
| 1       | Maximum impurity (binary classification)    |


### Gini Index

Another measure of impurity used in Decision Trees.

    Gini=1−∑pi^2
Where pi is the probability of class i.

#### Gini Range

| Gini         | Meaning                                    |
| ------------ | ------------------------------------------ |
| 0            | Pure node                                  |
| Close to 0.5 | Highly impure node (binary classification) |


### Entropy vs Gini

| Feature             | Entropy               | Gini Index     |
| ------------------- | --------------------- | -------------- |
| Formula             | (-\sum p_i\log_2 p_i) | (1-\sum p_i^2) |
| Uses Logarithm      | Yes                   | No             |
| Computational Speed | Slower                | Faster         |
| Used In             | ID3, C4.5             | CART           |
| Impurity Measure    | Yes                   | Yes            |
