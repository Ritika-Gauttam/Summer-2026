```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data
```

```python
plt.plot(data['total_bill'])
plt.show()
```

```python
sns.lineplot(data['total_bill'])
plt.show()
```

```python
plt.scatter(data['total_bill'],data['tip'])
```

```python
sns.scatterplot(x = 'total_bill',y= 'tip' ,data= data)
plt.show()
```

```python
plt.hist(data['total_bill'])
plt.show()
```

```python
sns.histplot(data['total_bill'])
plt.show()
```

```python
sns.kdeplot(data['total_bill'])
```

```python
# catplot and replot replot for relation
```

```python
sns.catplot(x='total_bill', y ='tip',data = data)
plt.show()
```

```python
sns.relplot(x='total_bill', y ='tip',data = data)
plt.show()
```

```python
sns.relplot(x='total_bill', y ='tip',hue = 'time',data = data)
```

```python
sns.relplot(x='total_bill', y ='tip',hue = 'smoker',data = data)
```

```python
sns.relplot(x='total_bill', y ='tip',hue = 'time', kind = 'scatter',col='smoker',data = data)        #kind can be of either line or scatter
```

```python
sns.relplot(x='total_bill', y ='tip',hue = 'time', kind = 'scatter',row='smoker',data = data)
```

```python
sns.relplot(x='total_bill', y ='tip',hue = 'time', kind = 'line',col='smoker',data = data)
```

```python
sns.relplot(x='total_bill', y ='tip',hue = 'time', kind = 'scatter',col='day',data = data)
plt.show()
```

```python
sns.relplot(x='total_bill', y ='tip',data = data)
plt.title("This is a scatter plot")
plt.xlabel("Total_bill",fontsize = 20 , color = "black")
plt.ylabel("Tips")
plt.show()
```

```python
sns.catplot(x = 'day',y = 'total_bill',kind= 'strip', data = data)
```

```python
sns.catplot(x = 'day',y = 'total_bill',kind= 'swarm', data = data)
```

```python
sns.catplot(x = 'day',y = 'total_bill',kind= 'bar', data = data)
```

```python
sns.catplot(x = 'day',y = 'total_bill',kind= 'box', data = data)
```

```python
sns.catplot(x = 'day',y = 'total_bill',kind = 'boxen', data = data)
```

```python
sns.catplot(x = 'day',y = 'total_bill',kind= 'point', data = data)
```
