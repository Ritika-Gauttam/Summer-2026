### LOOPING IN PYTHON:

* A loop in Python is used to repeat a block of code multiple times until a condition is satisfied.

        TYPES OF LOOP IN PYTHON

        Python mainly has 2 types of loops:

1. for loop

2. while loop


            1. for loop
* Used when the number of iterations is known.

### FOR LOOP 

* used for iterating one by one
* used when number of iterations is known

# Syntax
for x in variable_name:

        # code

```python
name = "Ritika"

for x in name:
    print(x)
    
```

```python
lst = [1,2,3,4]
for y in lst :
    print(y)    #### block of code 
    print (y+2)
```

### for loop with range()

* Syntax

1. for x in range(start , end):    ------> end is not included

       ##code

2. for x in range(start , end , step):    ------> end is not included

       ##code





```python
for x in range(1,10):
    print(x)
```

```python
for x in range(1,10,2):
    print(x)
```

```python
dict = {"name" : "Ritika" ,
        "branch" : "Aiml" ,
        "Roll no" : 18}

for x in dict:
    print(x) #------> by default with print keys
```

```python
for x in dict.keys():
    print(x)
```

```python
for x in dict.values():
    print(x)
```

```python
for x in dict.items():
    print(x)
```

```python
tpl =(1,3,4,5,6)
for item in tpl:
    print(item)
```

```python
lst = [1,2,3,5]

for x in lst:
    if x==3:
        break  #break is used to get out of loop
    print(x)
```

```python
lst = [1,2,3,5]

for x in lst:
    if x==3:
        continue      #continue is used to skip current
    print(x)
```

```python
fruits =["Apple","Banana","Grapes","Mango","Watermelon"]

for fruit in fruits:
    if fruit == "Mango":
        continue
    print(fruit)
```

```python
for x in range(1,20,2):
    if x==15:
        break
    if x==12:
        continue
    print(x)
```

```python
lst = [[2,3,45,56,5],[23,45,68,79,83]]  #we can store a list in a list
lst = [1,2,3,4,2,["Hello","hii","how"]]
print(lst[5])
print(lst[5][1])
```

# While loop in Python


Execute a block of code until a condition is True

Syntax:

      while Condition:
          ##Code



```python
x = 1
while x<=5 :
    print(x)
    x += 1
```

```python
x = 1
while x<=5 :
    print(x)
    x += 1
```

```python
x = 1
while x<=5 :
    print(x)
    x += 1
    if x==4:
        continue
    
```

```python
x = 1
while x<=5 :
    if x==4:
        break
    print(x)
    x += 1
    
    
```

```python
x = 5
while x>=1:
    print(x)
    x-=1


```
