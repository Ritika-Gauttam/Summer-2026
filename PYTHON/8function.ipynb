```python
lst = [12,34,23,45]

def even_odd(lst):
    for x in lst:
        if x % 2 == 0:
            print(x,"is an even number")
        else :
            print(x,"is an odd number")

even_odd(lst)

```

```python
def grade(marks):
    if marks>90 and marks<=100 :
        print("A++")
    elif marks <= 90 and marks > 80:
        print("A")
    elif marks <=80 and marks > 70:
        print("B")
    elif marks <= 70 and marks > 33:
        print("C")
    else :
        print("Fail")
    
grade(83)
```

```python
def prime_number(num):
    for x in range(2 , num):
        if num % x ==0:
            break
    

for x in range(1,100):
    print(prime_number(x))
```

```python
def prime():
    for x in range(1,101):
        if x > 1:
            for i in range(2,x):
                if x%i==0:
                    break
            else:
                print(x)

prime()
```
