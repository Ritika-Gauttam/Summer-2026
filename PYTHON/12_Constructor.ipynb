### Constructor


A constructor is a special method in Python that is automatically called when an object is created.

It is used to:
Initialize object data
Assign values to variables

**Syntax:**

    class Class_name:
        def __init__(self):
        pass

```python
class College:
    def __init__(self):
        pass
    
clg = College()
```

```python
class College:
    def __init__(self):
        college_name = "GWECA"
clg=College()
```

```python
class College:
    def __init__(self):
        self.college_name = "GWECA"     #instance variable
        self.college_address = "Ajmer"  #instance variable
        self.college_sp = "42"          #instance variable

clg = College()
```

```python
clg.college_name
```

```python
clg.college_address
```

```python
clg.college_sp
```

**Parameterized Constructor**

```python
class Student:
    def __init__(self, name , branch , roll_number , classes):
        self.name = name
        self.branch = branch
        self.roll_number = roll_number
        self.classes = classes
    
    def display(self):
        pass

clg = Student("Ritika","AIML", "23EEMAL022" , "3rd year")
```

```python
clg.classes
```

```python
clg.roll_number
```

```python
clg.display()
```

```python
class Student:
    def __init__(self, name , branch , roll_number , classes):
        self.name = name
        self.branch = branch
        self.roll_number = roll_number
        self.classes = classes
        
    def display(self):
        print(self.name)

clg = Student("Ritika","AIML", "23EEMAL022" , "3rd year")
clg.display()
```

MULTILEVEL BY CONSTRUCTOR

```python
class Vehicle:
    def __init__(self , brand ):
        self.brand= brand

class Car(Vehicle):
    def __init__(self ,brand, model):
        super().__init__(brand)
        self.model = model

class SportsCar(Car):
    def __init__(self , brand, model, speed):
        super().__init__(brand,model)
        self.speed = speed

    def display(self):
        print(self.brand)
        print(self.model)
        print(self.speed)

    
sc = SportsCar("mustang","M4","250")
sc.display()
```
