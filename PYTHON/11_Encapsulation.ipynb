## Encapsulation


Binding data (variables) and methods (functions) together inside a class and restricting direct access to some details.

**Types of Access Specifiers in Python**

| Access Type | Syntax        | Meaning                         |
| ----------- | ------------- | ------------------------------- |
| Public      | `self.name`   | Accessible everywhere           |
| Protected   | `self._name`  | Should not be accessed directly |
| Private     | `self.__name` | Strongly restricted             |





    Protected members use a single underscore _.

    Private members use double underscore __.

```python
class College:
    college_name = "GWECA"     ##public variable
    college_garden = "Yes"     ##public variable
    college_account = "233454932099409290"    ##public variable
```

```python
clg = College()
```

```python
clg.college_account
```

```python
clg.college_garden
```

```python
clg.college_name
```

```python
class College:
    college_name = "GWECA"     ##public variable
    college_garden = "Yes"     ##public variable
    __college_account = "233454932099409290"    ##private variable

clg1 = College()
```

```python
clg1.__college_account   #private member cannot be access by objects directly
```

```python
clg1.college_garden
```

```python
clg1.college_name
```

```python
class College:
    college_name = "GWECA"     ##public variable
    _college_garden = "Yes"     ##protected variable
    __college_account = "233454932099409290"    ##private variable

clg2 = College()
```

```python
clg2.college_name
```

```python
clg2._college_garden    #protected members can be accessed by object but we should not do this
```

```python
clg2.__college_account
```

#### Accessing private variables 

Method 1

```python
class College:
    college_name = "GWECA"     ##public variable
    _college_garden = "Yes"     ##protected variable
    __college_account = "233454932099409290"    ##private variable

    def private_variable_access(self):
        return self.__college_account

clg2 = College()
clg2.private_variable_access()
```
