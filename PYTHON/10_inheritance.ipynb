## Inheritance in Python

Inheritance is an Object-Oriented Programming (OOP) feature in which one class acquires the properties and methods of another class.

The existing class is called the Parent/Base/Super class
The new class is called the Child/Derived/Sub class


**Basic Syntax**


    class Parent:
        def show(self):
            print("Parent class method")

    class Child(Parent):
        pass

    obj = Child()
    obj.show()

Types of Inheritance in Python

| Type of Inheritance      | Definition                                                        | Structure       | Example                       |
| ------------------------ | ----------------------------------------------------------------- | --------------- | ----------------------------- |
| Single Inheritance       | One child class inherits from one parent class                    | A → B           | `class B(A)`                  |
| Multiple Inheritance     | One child class inherits from multiple parent classes             | A, B → C        | `class C(A, B)`               |
| Multilevel Inheritance   | A class inherits from a class which itself inherits another class | A → B → C       | `class C(B)`                  |
| Hierarchical Inheritance | Multiple child classes inherit from one parent class              | A → B, C        | `class B(A)` and `class C(A)` |
| Hybrid Inheritance       | Combination of two or more inheritance types                      | Mixed structure | Combination of above          |


### Single Inheritance

```python
class College:
    college_name = "GWECA"
    college_address = "jaipur"
    college_sp = 42

    def show(self):
        print(self.college_name)
        print(self.college_address )
        print(self.college_sp)

class Student(College):
    pass


st = Student()
st.show()
```

```python
# child class can also have its properties
class College:
    college_name = "GWECA"
    college_address = "jaipur"
    college_sp = 42

    def show(self):
        print(self.college_name)
        print(self.college_address )
        print(self.college_sp)

class Student(College):
    student_name = "Ritika"
    student_rollnumber = 18888
    student_number = 9245324432
    student_branch = "AIML"

st = Student()
st.show()
print(st.student_branch)  #We can access parent as well as child class properties thorugh a single child class object
```

### Multilevel Inheritance

```python
class Person:
    pass
class Teacher(Person):
    pass
class HOD(Teacher):
    pass
```

```python
class Person:
    def person_info(self):
        print("I am person")
class Teacher(Person):
    def teacher_info(self):
        print("I am Teacher")
class HOD(Teacher):
    def hod_info(self):
        print("I am HOD")

h = HOD()
print(h.hod_info())
print(h.person_info())
print(h.teacher_info())


# OUTPUT
# I am HOD
# None   -----> None because function is returning nothing
# I am person
# None   -----> None because function is returning nothing
# I am Teacher
# None   -----> None because function is returning nothing
```

### Hybrid Inheritance

```python
class Company1:
    pass
class Company2:
    pass
class Company3(Company1,Company2):
    pass
class Company4(Company3):
    pass
```

```python
class College:
    college_name = "GWECA"
    college_Address = "Ajmer"
    def college_info(self):
        print(self.college_name)
        print(self.college_Address)

class Student1(College):

    def student1_intro(self):
        print("I am student1")

class Student2(College):
    def student2_intro(self):
        print("I am student 2")

class Branch(Student1,Student2):
    def branch(self):
        print("My branch is AIML")

```
