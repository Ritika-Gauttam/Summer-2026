```python
class ATM_MACHINE:
    balance = 500
    def __init__(self):
        pass

    def deposit(self, amount):
        if(amount > 0):
            self.balance += amount
            print(amount , "desposited successfully")
        else:
            print("Please deposit a valid amount") 

    def withdraw(self,amount):
        if(self.balance >= amount):
            print(amount , "withdrawn successfully")
            self.balance -=amount
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print("Your current balance is Rs.",self.balance)

obj1 = ATM_MACHINE()
obj1.deposit(5000)
obj1.withdraw(2000)
obj1.check_balance()
```
